```python
# get a list of all of the folders in the data folder
import os
import pandas as pd
import matplotlib.pyplot as plt

folders = []
csv_files = []
for dirs in os.listdir('data'):
    # return the file that has the csv extension
    for file in os.listdir('data/' + dirs):
        if file.endswith('.csv'):
            folders.append(dirs)
            csv_files.append(file)
            break

# read the csv files into a dataframe

data = []

for i in range(len(csv_files)):
    data.append(pd.read_csv('data/' + folders[i] + '/' + csv_files[i]))

# get the column names
column_names = []
for i in range(len(data)):
    column_names.append(data[i].columns)

# print the column called 'key_resp.keys'
for i in range(len(data)):
    print(data[i]['key_resp.keys'])
```
